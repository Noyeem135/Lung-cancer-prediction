Data Pre-Processing

```python
df=df.drop(["index","Patient Id","Age"],axis=1)
```

```python
df.head()
```

```python
df["Level"].unique()
```

Appling replace_method

```python
df["Level"].replace({"Low":0, "Medium":1, "High":2},inplace=True)
```

```python
df.head()
```

```python
df["Level"].value_counts(ascending=True)
```

```python
plt.figure(figsize=(7,5))
sns.countplot(x="Level",order=[0,1,2,3],hue="Level",data=df).set_title("Level [Low,Medium,High]",fontsize=15,fontdict={"weight": "bold"})
```

```python
sns.displot(data=df, x="Level", height=5, aspect=2)
```

```python
df["Level"].value_counts(ascending=True)
```

```python
# Dividing dependent and independent variables
X = df.drop("Level", axis=1)
y = df["Level"]
```

```python
# Sampling
from imblearn.over_sampling import SMOTE 
from collections import Counter

sampler = SMOTE()
X, y = sampler.fit_resample(X, y)

print('SMOTE {}'.format(Counter(y)))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

Machine Learning

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression  

classifier= LogisticRegression(random_state=0)  
lr_model=classifier.fit(X_train, y_train)
y_pred=lr_model.predict(X_test)
```

```python
# Accuracy = TP+TN/TP+FP+FN+TN
# Precision = TP/TP+FP
# Recall = TP/TP+FN
#F1 Score = 2*(Recall * Precision) / (Recall + Precision)


from sklearn.metrics import classification_report, confusion_matrix 
print(classification_report(y_test, y_pred))
```

```python
x_axis_labels = ['Not Exit','Exit']
y_axis_labels = ['Not Exit','Exit']

confusion_m=confusion_matrix(y_test,y_pred)
sns.heatmap(confusion_m, annot=True,cmap='Greens', xticklabels=x_axis_labels, yticklabels=y_axis_labels)

plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')
```

Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
dt_model = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = dt_model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
x_axis_labels = ['Not Exit','Exit']
y_axis_labels = ['Not Exit','Exit']

confusion_m=confusion_matrix(y_test,y_pred)
sns.heatmap(confusion_m, annot=True,cmap='Greens', xticklabels=x_axis_labels, yticklabels=y_axis_labels)

plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')
```

KNN

```python
from sklearn.neighbors import KNeighborsClassifier

clf= KNeighborsClassifier(n_neighbors=25)

# Train the model using the training sets
knn_model=clf.fit(X_train,y_train)

y_pred=knn_model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
x_axis_labels = ['Not Exit','Exit']
y_axis_labels = ['Not Exit','Exit']

confusion_m=confusion_matrix(y_test,y_pred)
sns.heatmap(confusion_m, annot=True,cmap='Greens', xticklabels=x_axis_labels, yticklabels=y_axis_labels)

plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=100,random_state=0)
rf_model=clf.fit(X_train, y_train)
y_pred=rf_model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
x_axis_labels = ['Not Exit','Exit']
y_axis_labels = ['Not Exit','Exit']

confusion_m=confusion_matrix(y_test,y_pred)
sns.heatmap(confusion_m, annot=True,cmap='Greens', xticklabels=x_axis_labels, yticklabels=y_axis_labels)

plt.xlabel('Predicted Label')
plt.ylabel('Actual Label')
```

Support vector machine

```python
from sklearn.svm import LinearSVC as svc

svm_model = svc().fit(X_train, y_train)
y_pred=svm_model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

XG Boost

```python
from xgboost import XGBClassifier


# fit model no training data
clf = XGBClassifier(
              learning_rate=0.300000012,n_estimators=100, n_jobs=4, num_parallel_tree=1, random_state=0,
             )
xgboost_model=clf.fit(X_train, y_train)

y_pred=xgboost_model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```
